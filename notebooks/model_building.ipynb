# Part 4: Building a Model

## Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
sns.set_theme()
```

## Read data

```python
dirname = '../data/'
df = pd.read_csv(dirname + 'stations_pois.csv', sep=',')
df.head()
```

## Visualizations

```python
# Set dir to save images to.
dirname = '../images/'
```

```python
cols = df.columns.to_list()[2:]
nrows = 4
ncols = 2
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(7, 9))
for idx in range(len(cols)):
    r, c = idx // ncols, idx % ncols
    sns.scatterplot(data=df, x=cols[idx], y='free_bikes', ax=axes[r, c])
fig.tight_layout()
plt.savefig(dirname + 'scatter_plots.png')
```

```python
corr_matrix = df.loc[:, 'free_bikes':].corr()
sns.heatmap(corr_matrix, cmap='Purples', annot=True, fmt='.2f')
plt.savefig(dirname + 'heatmap.png')
```
